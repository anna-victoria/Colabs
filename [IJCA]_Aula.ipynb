```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df['sex'].value_counts().plot(kind='barh', color=['blue', 'm'], rot=0)
plt.title('Contagem de Passageiros por Gênero')
plt.ylabel('Gênero')
plt.xlabel('Número de Passageiros')
plt.show()
```

```python
df['sex'].value_counts().plot(kind='pie', autopct='%.1f%%')
plt.title('Contagem de Passageiros por Gênero')
plt.show()
```

```python
# plt.figure(figsize=(8, 6))
df['age'].plot(kind='hist', bins=40, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idades dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
df.plot.scatter(x='age', y='fare')
plt.title('Relação entre Idade e Tarifa')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.show()
```
